### Train test split on scaled data

```python
x = scaled_data.drop(['labels'],axis = 1)
y = scaled_data['labels']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# By the gridsearchCV

Scores = []
for model_name ,model_param in model_and_parameter.items():
    gs = GridSearchCV(model_param['model'],model_param['param'],cv = 5,return_train_score=False)
    gs.fit(x_train,y_train)
    Scores.append({'Model': model_name,
                 'Best_Score' : gs.best_score_,
                 'Best_Parameter' : gs.best_params_})
    best_model_of_algo = gs.best_estimator_
    joblib.dump(best_model_of_algo,'models/ML_models/'+model_name+str(int(round(gs.best_score_,2)*100))+'.lb')     # to save the best model of algorithm    
```

```python
model_info = pd.DataFrame(Scores,columns=['Model','Best_Score','Best_Parameter'])
model_info.sort_values(by='Best_Score',ascending=False)
```
